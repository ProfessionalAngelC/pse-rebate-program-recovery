```python
import matplotlib.pyplot as plt
```

```python
estimator_totals = combined.groupby('estimator')['amount'].sum().sort_values(ascending=False)

estimator_totals.plot(kind='bar', title='Total Rebate Value by Estimator (Other Utility Programs)')
plt.ylabel('Total Amount ($)')
plt.xlabel('Estimator')
plt.show()
```

## Summary

This notebook cleans and combines real rebate-program tracking data across 7 smaller utility programs (Cascade Gas, Tacoma Power, PUD, Clallam PUD, Lewis PUD, SnoPUD, Cowlitz PUD) — a complement to a separate, larger analysis of the PSE rebate program.

Key steps:
- Loaded 9 separate spreadsheet tabs, each with inconsistent column names, formatting, and structure
- Standardized column names across all tabs, fixed typos, removed empty/junk columns
- Extracted a structured status field from a multi-stage approval process (Clallam PUD 25)
- Repaired and validated malformed dates
- Combined all tabs into a single clean dataset (38 records)
- Anonymized individual names to match the cohort labels used in the related PSE analysis
- Verified no duplicate records after combining
- Produced summary aggregations and a visualization of rebate value by estimator

Total combined value: **$167,514.55** across all 7 programs, with PUD contributing the largest share ($91,183.00).
